```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv("/content/dataoodata.csv")
df1.head()
```

```python
df1.isna().sum()
```

```python
# Menghapus kolom 'user_id' karena tidak digunakan
df1.drop(columns=['user_id'], inplace=True)
print(df1.head())
```

```python
# Mengatur ukuran gambar
plt.figure(figsize=(10, 6))

# Membuat heatmap
sns.heatmap(df1.corr(), annot=True, cmap='Reds', fmt='.2f')
plt.title('Heatmap Dataset Customer Purchasing Behaviors')
plt.show()
```

Tujuan, User, Fungsi dan Tone:
### 1. Tujuan
Menyediakan Wawasan Korelasi: Memvisualisasikan hubungan antar variabel untuk memahami pola perilaku pembelian pelanggan.

### 3. Fungsi
- Visualisasi Hubungan: Menunjukkan kekuatan hubungan antar variabel numerik secara jelas.
- Mendukung Pengambilan Keputusan: Memberikan dasar untuk keputusan berbasis data.
